```python
import os
import torch
import shutil
import cv2
import numpy as np
from PIL import Image
```

```python
path = "./datasets_segmentation/brain_mri"
dir_list = os.listdir(path)

images = []
# Iterate over all the directories
for dir_name in dir_list:
    dir_path = path + '/' + dir_name
    if os.path.isdir(dir_path):
        for image in os.listdir(dir_path):
            if not image.endswith('_mask.tif'):
                i = dir_path + '/' + image
                m = dir_path + '/' + image[:-4] + '_mask.tif'
                images.append([i, m])
                
                
out_path = "/media/jonas/Seagate Expansion Drive/Memoria/master_thesis"
# Split
# train_data, val_data, test_data, discard = torch.utils.data.random_split(images, [15, 15, 15, 3929 - 45])
train_data, val_data, test_data = torch.utils.data.random_split(images, [3000, 900, 29])

os.makedirs(f'{out_path}/datasets_segmentation/Brain/train')
os.makedirs(f'{out_path}/datasets_segmentation/Brain/val')
os.makedirs(f'{out_path}/datasets_segmentation/Brain/test')

for data in test_data:
    im_name0 = data[0].split('/')[-1]
    im_name1 = data[1].split('/')[-1]
    im0 = cv2.imread(data[0])
    im1 = cv2.imread(data[1])
    im0 = im0 / 255
    im1 = im1 / 255
    im0 = im0.transpose(2, 0, 1)
    im1 = np.dot(im1[...,:3], [0.299, 0.587, 0.114])
    np.save(f'{out_path}/datasets_segmentation/Brain/test/{im_name0[:-4]}', im0)
    np.save(f'{out_path}/datasets_segmentation/Brain/test/{im_name1[:-4]}', im1)
#     img = Image.fromarray(gim)
#     img.show()
#     break

for data in train_data:
    im_name0 = data[0].split('/')[-1]
    im_name1 = data[1].split('/')[-1]
    im0 = cv2.imread(data[0])
    im1 = cv2.imread(data[1])
    im0 = im0 / 255
    im1 = im1 / 255
    im0 = im0.transpose(2, 0, 1)
    im1 = np.dot(im1[...,:3], [0.299, 0.587, 0.114])
    np.save(f'{out_path}/datasets_segmentation/Brain/train/{im_name0[:-4]}', im0)
    np.save(f'{out_path}/datasets_segmentation/Brain/train/{im_name1[:-4]}', im1)
    

for data in val_data:
    im_name0 = data[0].split('/')[-1]
    im_name1 = data[1].split('/')[-1]
    im0 = cv2.imread(data[0])
    im1 = cv2.imread(data[1])
    im0 = im0 / 255
    im1 = im1 / 255
    im0 = im0.transpose(2, 0, 1)
    im1 = np.dot(im1[...,:3], [0.299, 0.587, 0.114])
    np.save(f'{out_path}/datasets_segmentation/Brain/val/{im_name0[:-4]}', im0)
    np.save(f'{out_path}/datasets_segmentation/Brain/val/{im_name1[:-4]}', im1)
    
    
    
    
    
    
# for data in train_data:
#     im_name0 = data[0].split('/')[-1]
#     im_name1 = data[1].split('/')[-1]
#     im0 = cv2.imread(data[0])
#     im1 = cv2.imread(data[1])
#     im0 = im0 / 255
#     im1 = im1 / 255
#     im0 = np.transpose(im0, [2, 0, 1])
#     im1 = np.transpose(im1, [2, 0, 1])
#     np.save(f'datasets_segmentation/Brain/train/{im_name0[:-4]}', im0)
#     np.save(f'datasets_segmentation/Brain/train/{im_name1[:-4]}', im0)
    
    
# for data in test_data:
#     im_name0 = data[0].split('/')[-1]
#     im_name1 = data[1].split('/')[-1]
#     shutil.copyfile(data[0], f'datasets_segmentation/Brain/test/{im_name0}')
#     shutil.copyfile(data[1], f'datasets_segmentation/Brain/test/{im_name1}')
    
# for data in val_data:
#     im_name0 = data[0].split('/')[-1]
#     im_name1 = data[1].split('/')[-1]
#     shutil.copyfile(data[0], f'datasets_segmentation/Brain/val/{im_name0}')
#     shutil.copyfile(data[1], f'datasets_segmentation/Brain/val/{im_name1}')
```
